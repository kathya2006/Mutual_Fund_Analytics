```python
df.drop_duplicates(inplace=True)
df.fillna(df.mean(numeric_only=True),inplace=True)
```

```python
print(df.columns)
```

```python
df['Date'] = pd.to_numeric(df['Date'])
```

```python
df['Sale Price'] = pd.to_numeric(df['Sale Price'])
```

```python
top = df.sort_values('Sale Price', ascending=False).head(10)
top
```

```python
df.groupby('category')['Sale Price'].mean()
```

```python
df.groupby('Net Asset Value')['Sale Price'].mean()
```
